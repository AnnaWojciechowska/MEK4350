```python
import sympy as sp
from sympy import symbols, sqrt, exp, pi, init_printing, cancel, latex, simplify
x = symbols('x')
import numpy as np
from scipy.stats import skew, kurtosis
import matplotlib.pyplot as plt
```

```python
x,alpha = symbols('x alpha')
with sp.evaluate(False):
    Rayleigh = x/alpha**2 * exp(-(x**2)/(2*alpha**2))
Rayleigh
```

```python
Rayleigh_fun = sp.lambdify((x, alpha), Rayleigh, 'numpy')
x_values = np.arange(0,5, 0.05)
fig, ax = plt.subplots(figsize=(25, 5))
alpha_value = 1
y_values_Rayleigh = Rayleigh_fun(x_values, alpha_value)
_ = ax.plot(x_values, y_values_Rayleigh)
```

```python
fig, ax = plt.subplots(figsize=(25, 5))
for alpha_value in [0.3, 0.5, 1, 2, 5]:
    y_values_Rayleigh = Rayleigh_fun(x_values, alpha_value)
    _ = ax.plot(x_values, y_values_Rayleigh)
```

```python
x,alpha = symbols('x alpha')
with sp.evaluate(False):
    Laplace = 1/2*alpha * exp(-abs(x)/alpha)
Laplace
```

```python
Laplace_fun = sp.lambdify((x, alpha), Laplace, 'numpy')
x_values = np.arange(-5,5, 0.05)
fig, ax = plt.subplots(figsize=(25, 5))
alpha_value = 0.1
y_values_Laplace = Laplace_fun(x_values, alpha_value)
_ = ax.plot(x_values, y_values_Laplace)
```

```python
fig, ax = plt.subplots(figsize=(25, 5))
for alpha_value in [0.3, 0.5, 1, 2, 5]:
    y_values_Laplace = Laplace_fun(x_values, alpha_value)
    _ = ax.plot(x_values, y_values_Laplace)
```
