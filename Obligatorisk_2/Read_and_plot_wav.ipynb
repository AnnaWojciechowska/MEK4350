```python
import wave
import numpy as np
import matplotlib.pyplot as plt
```

https://docs.python.org/3/library/wave.html

```python
file_path = '/home/anna/annaCode/UiO/MEK4350/notebooks/oblig2_data/case1.wav'
with wave.open(file_path, 'rb') as wf:
    frames = wf.readframes(-1)
    signal = np.frombuffer(frames, dtype=np.int16)
    framerate = wf.getframerate()
    time = np.linspace(0, len(signal) / framerate, num=len(signal))
```

```python
print(f"The sampling frequnecy is {framerate} Hz or {framerate/1000} KHz")
```

```python
print(f"{len(time):.2e}", f"{len(frames):.2e}", f"{len(signal):.2e}",)
```

```python
len(time), len(frames), len(signal)
```

```python
n_secs = 0.01 # [s]
n_secs_points = int(n_secs * framerate) 
n_secs_points
fig, ax = plt.subplots(figsize=(25, 5))
_ = ax.plot(time[:n_secs_points], signal[:n_secs_points])
```
