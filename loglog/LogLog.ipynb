```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
x = np.linspace(0.1, 2, 400)  # You can adjust the range and number of points as needed
a = 2
b = 3
y = a*x**b
```

```python
plt.figure(figsize=(8, 6)) 
plt.plot(x, y, label=f'y = {a}x^{b}', color='b', linewidth=2)  
plt.xlabel('x')  # X-axis label
plt.ylabel('y')  # Y-axis label
plt.title('Plot of y = ax^b')  # Plot title
plt.legend()  
plt.grid(True)  
plt.show() 
```
