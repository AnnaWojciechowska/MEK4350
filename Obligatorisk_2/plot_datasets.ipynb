```python
from dataset import *
import os
```

```python
data_dir = './oblig2_data'
data_file_names = ['proberun4_tett.csv', 'BayOfBiscay.mat','New_Year_Wave.txt','case1.wav']
dataset_types = [ExperimentDataset, HDF5_Dataset, DaupnerDataset,WaveDataset]
```

```python
datasets = []
for i in range(0,4):
    filepath_dataset = os.path.join(data_dir, data_file_names[i])
    datasets.append(dataset_types[i](filepath_dataset))
    
    print(f"Data set: {i+1}")
    print(f"number of samples: {datasets[i].number_of_samples()}")
    print(f"sampling rate: {round(datasets[i].sample_rate,3)} [Hz]")
    print(f"Problem I b) Significant wave height Hs: {round(datasets[i].SWH(),3)}[m]")
    print(f"Problem I b) mean μ: {round(datasets[i].mean(),5)}[m]")
 
    print(f"Problem I c) Frequency resolution ∆f: {round(datasets[i].frequency_resolution,4)}[Hz]")
    print(f"Problem I c) Nyquist frequency: {round(datasets[i].Nyqyist_frequency(),2)} [Hz]")
    
    
    print()
```

```python
for i in range(0,4):
    datasets[i].plot_data()
```
